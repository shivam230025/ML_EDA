# **EDA-3**- Amazon Sales Data

```python
df= pd.read_csv('amazon.csv')
df.head()
```

**1- What is the average rating for each product category?**

```python
# dropping row with some goofy rating value - '|'
df= df[df['rating']!='|']

# changing rating column to float type
df['rating']= df['rating'].astype('float')

# doing feature engineering on category column
# in truth category column is showing whole path from category to the product
# to extract catagory we just need main category ie. fist string value before |

def categoryFE(category):
  return category.split('|')[0]

df['category_upd']= df['category'].apply(categoryFE)
```

```python
df['category_upd'].value_counts()
```

```python
df.groupby('category_upd')['rating'].mean().round(2)
```

**2- What are the top rating_count products by category?**

```python
# removing ',' from the numerical values
df['rating_count']= df['rating_count'].str.replace(',','')

# changing data type to numerical
df['rating_count']= df['rating_count'].astype('float')
```

```python
# getting ids of the top rating_count products
idx= df.groupby('category_upd')['rating_count'].idxmax()

# showing top rating_count product name corresponding to each category
df.loc[idx, ['category_upd', 'product_name']].set_index('category_upd')
```

**3- What is the distribution of discounted prices vs. actual prices?**

```python
# function to remove rupees sign
def removeRupeesSign(amount):
  return amount[1:]

df['actual_price']= df['actual_price'].str.replace(',','')
df['actual_price']= df['actual_price'].apply(removeRupeesSign)
df['actual_price']= df['actual_price'].astype('float')

df['discounted_price']= df['discounted_price'].str.replace(',','')
df['discounted_price']= df['discounted_price'].apply(removeRupeesSign)
df['discounted_price']= df['discounted_price'].astype('float')
```

```python
# we can use kde plot
plt.figure(figsize=(10,5))
col= ['actual_price', 'discounted_price']

for c in col:
  sns.kdeplot(df[c], fill=True, label=f'{c}')
plt.xlabel('Prices')
plt.legend()
plt.show()
```

**4- How does the average discount percentage vary across categories?**

```python
# removing % from the last
df['discount_percentage']= df['discount_percentage'].apply(lambda x: x[:-1])
```

```python
# changing column type to int
df['discount_percentage']= df['discount_percentage'].astype('int')
```

```python
# dispalying average discount percentage vary across categories
print(df.groupby('category_upd')['discount_percentage'].mean().round(2))
df.groupby('category_upd')['discount_percentage'].mean().round(2).plot(kind='bar', xlabel='category', ylabel='average discount %')
```

**5- What are the most popular product names?**

```python
# with most popular meaning the products with most rating count
df1= df[['rating_count', 'product_name']]
df1.sort_values('rating_count', ascending=False, inplace=True)

# Top 20 most popular product names
df1.head(20)
```

**6- What are the most popular product keywords?**

```python
import re
from collections import Counter

# getting product list
product_list= df['product_name'].str.lower().to_list()

words=[]
for name in product_list:
  # cleaning product name of any special character and replacing it with ' '
  clean= re.sub(r'[^\w\s]',' ', name)
  clean_words= clean.split()

  # filtering stop_words, word with length less than 4, and any digit
  stop_words = {'with'}
  filtered= [w for w in clean_words if len(w)>3 and w not in stop_words and not w.isdigit()]

  words.extend(filtered)

# counting each word
word_counts= Counter(words)

# creating data frame for top 20 key words
top_keywords= pd.DataFrame(
    word_counts.most_common(20),
    columns= ['popular keyword', 'count']
)

print(top_keywords)
```

**9- Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.**

```python
# as there are multiple artists so taking top 10 artists with highest duration sum
top_10= df.groupby('Artist')['Duration (ms)'].sum().sort_values(ascending=False).head(10).index
# df containing only top 10 artists data
df1= df[df['Artist'].isin(top_10)]

plt.figure(figsize=(10,5))
sns.boxplot(data=df1, x='Artist', y='Duration (ms)', palette='viridis')
plt.xticks(rotation=45)
plt.title("Duration of tracks vary with different artists(Top 10)")
plt.show()
```

**10- How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.**

```python
# as there are multiple artists so taking top 10 artists with highest duration sum
top_10= df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False).head(10).index
# df containing only top 10 artists data
df1= df[df['Artist'].isin(top_10)]

plt.figure(figsize=(10,5))
sns.violinplot(data=df1, x='Artist', y='Popularity', palette='viridis')
plt.xticks(rotation=45)
plt.title("popularity of tracks vary with different artists(Top 10)")
plt.show()
```
